```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import dataOf12 as d


fig, ax = plt.subplots()

## 4*19*19
x=np.zeros(19)
data=np.zeros(19)
for l in range(1): ## l -> web
    for m in range(19): ## m->stiff
        i=m
        j=l*19*3
        k=19*19 ## thickness = 10
    
        x[m]=d.dispDataOfUpperRight[i+j+k][2]/300
        data[m] = d.dispDataOfDownRight[i+j+k][3]-d.dispDataOfUpperRight[i+j+k][3] #-dispDataOfUpperRight[yLine*15+i][2]
    ax.plot(x, data,label='Change the height of stringers')
for l in range(1): ## l -> web
    for m in range(19): ## m->stiff
        i=m*19
        j=l*3
        k=19*19 ## thickness = 10
      
        x[m]=d.dispDataOfUpperRight[i+j+k][1]/1200
        data[m] = d.dispDataOfDownRight[i+j+k][3]-d.dispDataOfUpperRight[i+j+k][3] #-dispDataOfUpperRight[yLine*15+i][2]
    ax.plot(x, data,label='Change the height of beams')


ax.set_xlabel('Multiplier of the height change')
ax.set_ylabel('Displacement(mm)')
plt.title(r'Displacement (U3) $D_2 - D_1$')
plt.legend()
#plt.show()
plt.savefig("/Users/b1kim/Desktop/MDPI_template/figures/trendMulti")
```
